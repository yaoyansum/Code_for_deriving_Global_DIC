```python
# 导入必要库
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

# 初始化结果列表

for j in range(16,18):
    results = []
    num = 50000 * (j - 1) + 1
    for i in range(num, num+50000):
        try:
            # 1. 数据加载
            data = pd.read_csv(f"G:\\流域属性\\模型二湖泊分类2\\{i}.0.csv")
            X = data.iloc[:, 2:]
            y = data["DIC"]

            # 2. 训练模型
            model = RandomForestRegressor(
                n_estimators=200,
                random_state=42,
                max_depth=8,
                n_jobs=2  # 使用所有CPU核心
            )
            model.fit(X, y)

            # 3. 提取特征重要性
            feature_names = X.columns
            importances = model.feature_importances_
            feature_importance = pd.DataFrame({
                'Feature': feature_names,
                'Importance': importances
            }).sort_values('Importance', ascending=False)

            # 提取最高重要性特征
            top_feature = feature_importance.iloc[0]
            results.append({
                'i': i,
                'Feature': top_feature['Feature'],
                'Importance': top_feature['Importance']
            })

    #         # 4. 打印结果（可选）
    #         print("="*50)
            print(i)
    #         print("回归模型特征重要性：")
    #         print(feature_importance.head(1))
    #         print(f"\n模型R^2得分: {model.score(X, y):.4f}")
    #         print("="*50)

        except Exception as e:
            print("="*50)
            print(f"Error processing {i}: {str(e)}")
            print("="*50)

    # 5. 保存结果到CSV
    result_df = pd.DataFrame(results)
    result_df.to_csv(f"G:\\流域属性\\分类结果2\\feature_importance_results{j}.csv", index=False)
```
