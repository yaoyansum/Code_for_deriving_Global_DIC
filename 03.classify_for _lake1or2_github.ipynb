```python
# 
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

# 

for j in range(16,18):
    results = []
    num = 50000 * (j - 1) + 1
    for i in range(num, num+50000):
        try:
            # 1. 
            data = pd.read_csv(f"G:\\\\2\\{i}.0.csv")
            X = data.iloc[:, 2:]
            y = data["DIC"]

            # 2. 
            model = RandomForestRegressor(
                n_estimators=200,
                random_state=42,
                max_depth=8,
                n_jobs=2  # CPU
            )
            model.fit(X, y)

            # 3. 
            feature_names = X.columns
            importances = model.feature_importances_
            feature_importance = pd.DataFrame({
                'Feature': feature_names,
                'Importance': importances
            }).sort_values('Importance', ascending=False)

            # 
            top_feature = feature_importance.iloc[0]
            results.append({
                'i': i,
                'Feature': top_feature['Feature'],
                'Importance': top_feature['Importance']
            })

    #         # 4. 
    #         print("="*50)
            print(i)
    #         print("")
    #         print(feature_importance.head(1))
    #         print(f"\nR^2: {model.score(X, y):.4f}")
    #         print("="*50)

        except Exception as e:
            print("="*50)
            print(f"Error processing {i}: {str(e)}")
            print("="*50)

    # 5. CSV
    result_df = pd.DataFrame(results)
    result_df.to_csv(f"G:\\\\2\\feature_importance_results{j}.csv", index=False)
```
